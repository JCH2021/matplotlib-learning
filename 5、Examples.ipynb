```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(100)
people = ['Tom', 'Dick', 'Harry', 'Slim', 'Jim']
#官方数据
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))
fig, ax = plt.subplots()
hbars = ax.barh(y_pos, performance, xerr=error, align='center')
#水平柱图
# ax.set_yticks(y_pos,labels=people) 这句用的时候会报错，我也没弄明白
#或将上面两句改为这一句 ax.barh(people,performance)
ax.invert_yaxis()  
#翻转 y 轴，否则从上到下依次为 Jim，Slim，···，Tom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')
ax.bar_label(hbars, fmt='%.2f')
#每个柱子练出一条指示数值的线，以及数值的格式
ax.set_xlim(right=15)  
#设置 x 轴的右限
plt.show()
```

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
# ax.set_xticks(x, labels) 这句用的时候会报错，我也没弄明白
ax.legend()
ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
fig.tight_layout()
#可以通俗的理解为调整图像之间的距离
plt.show()
```

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
#从结果来看参数一是各成分比例，
#参数二是饼图向外的偏移，因为只有30%不是0
#参数四是数字格式
#参数五在饼下画阴影
#参数六是从x轴逆时针旋转饼图开始的角度
ax1.axis('equal') 
#参数很多，大家可以自己看看
plt.show()
```
