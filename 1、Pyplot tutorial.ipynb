```python
names = ['group_a', 'group_b', 'group_c']
values = [50, 10, 90]
plt.figure(figsize=(9, 3))
#整个图形的宽高
plt.subplot(131)
plt.bar(names, values)
#绘制柱状图，参数一是横轴标记，参数二是纵轴值
plt.subplot(132)
plt.scatter(names, values)
#绘制点图
plt.subplot(133)
plt.plot(names, values)
#绘制折线图
plt.suptitle('Categorical Plotting')
#绘制标题
# plt.show()
```

```python
def f(x):
    return 1/(1+np.exp(-x))
x = np.arange(0,10,1)
plt.plot(x,f(x),'mo')
#这里其实和上面一样，只不过将 y 用函数的形式表示
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(1000)
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)
#绘制直方图，具体什么意思我还没懂
plt.xlabel('Smarts')
# x 轴标签
plt.ylabel('Probability')
# y 轴标签
plt.title('Histogram of IQ')
#标题
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
#以坐标为(60,0.025)处为文本左下角绘制文本
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
#是否增加网格
# plt.show()
```

```python
ax = plt.subplot()
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)
plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )
# xy 为箭头，xytext 为文字左下角，参数四是字典类型黑箭头，参数四还有别的类型
plt.ylim(-2, 2)
#获取或设置当前轴的 y 限制，也就是上下限
plt.xlim(0, 1.5)
# plt.show()
```

```python
#下面是各种线型

np.random.seed(100)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))
plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)
# plt.show()
```
